```python
# Importamos los modulos de interés
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
# Creamos la base de datos
engine = create_engine('sqlite:///:memory:')
```

```python
# Ahora importaremos pandas
import pandas as pd
```

```python
# Tenemos nuestro csv en un repositorio github
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
# Creamos el df
df = pd.read_csv(url)
df.head()
```

```python
df.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
inspector.get_table_names()
```

```python
# Haciendo consultas
query1 = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query1, engine)
empregados.head()
```

```python
# Guardando resultado

empregados.to_sql('empleados', con=engine, index=False)
```

```python
# Leyendo la tabla guardada
pd.read_sql_table('empleados', engine)
```

```python
# Leyendo solo columnas de interes
pd.read_sql_table('empleados', engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual']).head()
```

```python
query2 = 'SELECT * FROM clientes'
pd.read_sql(query2, engine)
```

```python
# Vamos a borrar un dato da la tabla
# Comenzamos creando la consulta que solicitará eliminar el registro
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
```

```python
# Importamos el modulo "text" de sqlalchemy
from sqlalchemy import text

# Ejecutamos la consulta
with engine.connect() as conn:
    conn.execute(text(query))
    conn.commit()
```

```python
# Consutalmos nuevamente para verificar que fue eliminado el id 

pd.read_sql(query2, engine)
```

```python
# Vamos a crear un a consulta para actualizar grado de escolaridad de del cliente id=5008808
query3 = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
```

```python
# Ahora ejecutaremos la consulta
with engine.connect() as conn:
    conn.execute(text(query3))
    conn.commit()
```
