```python
#Ahora vamos trabajar importando datos de hojas de calculo
import pandas as pd
```

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'
```

```python
datos = pd.read_excel(url)
datos.head()
```

```python
#Para saber si el archivo tiene mas hojas con información usamos el siguiente codigo
pd.ExcelFile(url).sheet_names
```

```python
#Por defecto siempre abre la primera hoja
#Pero podemos especificar que hoja queremos abrir con el parametro sheet_name()
#Se puede usar el nombre de la hoja o el indice(el primero es igual a 0)
percapita = pd.read_excel(url, sheet_name='emissoes_percapita')
percapita.head()
```
